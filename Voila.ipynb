```
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import dlib
from imutils import face_utils
from google.colab.patches import cv2_imshow
```

```
face_cascade=cv.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade=cv.CascadeClassifier('haarcascade_eye.xml')
leye_cascade=cv.CascadeClassifier('haarcascade_lefteye_2splits.xml')
mouth_cascade=cv.CascadeClassifier('haarcascade_mouth.xml')
#nose_cascade=cv.CascadeClassifier('haarcascade_mcs_nose.xml')
```

```
original=cv.imread('face.jpg')
gray = cv.cvtColor(original, cv.COLOR_BGR2GRAY)
plt.figure(figsize=(12,8))
plt.imshow(gray, cmap='gray')
plt.show()
```

```
# Detect faces
faces = face_cascade.detectMultiScale(
gray,
scaleFactor=1.1,
minNeighbors=5,
flags=cv.CASCADE_SCALE_IMAGE
)
# For each face
for (x, y, w, h) in faces: 
    # Draw rectangle around the face
    cv.rectangle(original, (x, y), (x+w, y+h), (255, 255, 255), 3)
```

```
cv2_imshow(original)
```

```
# Detect eyes
eyes = eye_cascade.detectMultiScale(
gray,
scaleFactor=1.1,
minNeighbors=9,
flags=cv.CASCADE_SCALE_IMAGE
)
# For each eye
for (x, y, w, h) in eyes: 
    # Draw rectangle around the eye
    cv.rectangle(original, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```
cv2_imshow(original)
```

```
# Detect left eye
leye = leye_cascade.detectMultiScale(
gray,
scaleFactor=1.1,
minNeighbors=15,
flags=cv.CASCADE_SCALE_IMAGE
)

for (x, y, w, h) in eyes: 
    # Draw rectangle around the eye
    cv.rectangle(original, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```
# Detect mouth
mouth = mouth_cascade.detectMultiScale(
gray,
scaleFactor=1.1,
minNeighbors=70,
flags=cv.CASCADE_SCALE_IMAGE
)

for (x, y, w, h) in mouth: 
    # Draw rectangle around the mouth
    cv.rectangle(original, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
